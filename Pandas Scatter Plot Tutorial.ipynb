## How to make a Scatter Plot

This is code for the [Python Data Visualization tutorial on how to create scatter plots using Python](https://youtu.be/TexdD7t0IKU). 

```python
%matplotlib inline
import pandas as pd

# Suppress warnings
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('https://vincentarelbundock.github.io/Rdatasets/csv/datasets/mtcars.csv', index_col=0)
```

```python
df.head()
```

## How to use Pandas Scatter Method to Create a Scatter Plot
First scatter plot example using plot.scatter() to create a simple scatter plot

```python
df.plot.scatter(x='wt', y='mpg')
```

## How to use Seaborn to create Scatter Plots in Python

We will continue using the Python library to create scatter diagrams. This enables us to get the regression line on the scatter plot.

```python
import seaborn as sns

# "Styling"
sns.set_style('white')
sns.set_style('ticks')

# Using regplot to create the scatter plot with 95% confidence interval and a regression line
sns.regplot(x='wt', y='mpg', data=df)
```

## Changing the Color on the Scatterplot

```python
sns.regplot(x='wt', y='mpg', color='g', data=df)
```

## Creating a Scatter plot using Seaborn lmplot

```python
sns.lmplot(x='wt', y='mpg', data=df)
```

```python
sns.lmplot(x='wt', y='mpg', hue='vs', data=df)
```

```python
sns.lmplot(x='wt', y='mpg', hue='vs', palette="Set1", data=df)
```

```python
sns.lmplot(x='wt', y='mpg', hue='vs',palette="Set1", markers='+', data=df)
```

```python
sns.lmplot(x='wt', y='mpg', hue='vs',palette="Set1", markers=['+', 'o'], data=df)
```
