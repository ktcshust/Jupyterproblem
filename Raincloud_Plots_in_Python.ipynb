### How to Create Raincloud Plots in Python
In this python data visualization tutorial we are going to learn about a new way to visualize data; the raincloud plot. This type of plot combines the boxplot, violin plot, & the scatter plot into one visualization. This [tutorial](https://youtu.be/uFDdzTjn190) requires that you have ptitprince and it's dependencies installed. Let's start by installing it using pip:

```
pip install ptitprince
```



### Creating a Raincloud Plot using Ptitprince
Now that we have ptitprince installed we can start by importing pandas as pd and ptitprince as pt:

````python
%matplotlib inline
import pandas as pd
import ptitprince as pt


# Ignore Warnings
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
````

### Reading a CSV file from an URL
Next we are going to read a CSV file from an url. We use Pandas read_csv to do this and if you are looking for a good pandas read_csv tutorial, check the link in the description below.


````python
df = pd.read_csv('https://vincentarelbundock.github.io/Rdatasets/csv/datasets/iris.csv')
````

### Raincloud Data Visualization in Python
Now, we are ready to create our first Raincloud Plot using Python!

````python
ax = pt.RainCloud(x = 'Species', y = 'Sepal.Length', 
                  data = df, 
                  width_viol = .8,
                  width_box = .4,
                  figsize = (12, 8), orient = 'h',
                  move = .0)
````

### Adding a Pointplot Connecting the Means
We can also add a Pointplot connecting the means by changing the "pointplot" argument to True

````python
ax = pt.RainCloud(x = 'Species', y = 'Sepal.Length', 
                  data = df,
                  pointplot = True,
                  width_viol = .8,
                  width_box = .4,
                  figsize = (12, 8), orient = 'h',
                  move = .0)
````
