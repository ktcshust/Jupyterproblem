## Slice, index, select Pandas Dataframe

This is the code for the [YouTube Tutorial on using iloc and loc to slice pandas dataframe](https://youtu.be/W1ryq6Wvg8g)

```python
import pandas as pd

# Create some variables
trials = range(1, 21)
n_rows = len(trials)
subj_id = [1]*n_rows
group = ['Control']*n_rows
condition = ['Affect']*int((n_rows/2)) + ['Neutral']*int((n_rows/2))

# Create a dictionairy
data = {'Condition':condition, 'Subject_ID':subj_id, 
        'Trial':trials, 'Group':group}

# Create the dataframe
df = pd.DataFrame(data)
df.head(10)
```

```python
# Select row five to 10
df[4:10:1]
```

```python
df.loc[4:9,:]
```

```python
df.loc[10:14,:]
```

```python
df.loc[0, 'Condition']
```

```python
# Printing out one column
df.loc[:, 'Condition']
```

```python
# Printing out two columns
df.loc[:, ['Condition', 'Group']]
```

```python
# Printing out from second column
df.loc[4:9, ['Condition', 'Group']]
```

```python
# From column 'Subject_ID' and forward
df.loc[4:9, 'Subject_ID':]

# From column 'Trial' and forward
# df.loc[4:9, 'Trial':]
```

```python
# Printing out fifth to the twentieth row and all columns
# Note the difference!
df.iloc[4:10,:]
```

```python
df.iloc[4:10, 0]
```

```python
df.iloc[4:10, [0, 3]]
```

```python
df.iloc[4:10, 1:]
```
