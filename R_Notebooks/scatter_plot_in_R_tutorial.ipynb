## <span class="ez-toc-section" id="How_to_use_Ggplot2_to_Produce_Scatter_Plots_in_R">How to use Ggplot2 to Produce Scatter Plots in R</span>

In this section we will learn how to make scattergraphs in R using ggplot2.

### <span class="ez-toc-section" id="How_to_Make_a_Scatter_Plot_in_R-2">How to Make a Scatter Plot in R</span>

We will start by visualizing the variables <strong>wt </strong>(x-axis) and <strong>mpg </strong>(y-axis). 

Before going on and creating the first scatter plot in R we will briefly cover ggplot2 and the plot functions we are going to use. First, we start by using <em>ggplot</em> to create a plot object.

Inside of the <code>ggplot()</code> function, we’re calling the <code>aes()</code> function that describe how variables in our data are mapped to visual properties . In this simple scatter plot in R example, we only use the x- and y-axis arguments and ggplot2 to put our variable wt on the x-axis, and put mpg on the y-axis. 

```r
require(ggplot2)

gp <- ggplot(aes(x = wt, y = mpg), 
             data = mtcars)

gp + geom_point()
```

### <span class="ez-toc-section" id="How_to_Change_the_Size_of_the_Dots_in_a_Scatter_Plot">How to Change the Size of the Dots in a Scatter Plot</span>

Here we'' change the size of the markers size using the <em>size</em> argument.

```r
gp + geom_point(size = 4)
```

Note, we used <code>aes()</code> but added the size argument to the <code>geom_point()</code> function. 

```r
gp + geom_point(aes(size = wt))
```

#### How to Change the Number of ´ticks using ggplot2

To change the x-axis we use the function <code>scale_x_continuous</code> and to change the y-axis we use the function <code>scale_y_continuous</code>. Furthermore, we use the arguments limits, which take a vector, and we can set the limits to change the ticks. 

```r
gp <- ggplot(aes(x = wt, y = mpg), 
             data = mtcars) + 
    geom_point()

gp + scale_y_continuous(limits=c(1, 40)) +
   scale_x_continuous(limits=c(0, 6))
```

Next we also change the number of ticks by adding the breaks argument to the above functions. Furthermore, we add the <code>seq</code> function to create a numeric vector.

```r
gp + scale_y_continuous(limits=c(1, 35),
                                breaks=seq(1, 35, 5)) +
   scale_x_continuous(limits=c(1.5, 5.5),
                     breaks=seq(1.5, 5.5, 1))
```

#### Grouped Scatter Plot in R

Here we group by using <em>color </em>argument and the <code>factor</code> function to change the variable vs to a factor.

```r
gp <- ggplot(aes(x=wt, y=mpg, color=factor(vs)), 
             data=mtcars)
gp + geom_point()
```

Another option is using the <code>as.factor</code> function and change *vs* to a factor in the dataframe object.

```r
mtcars$vs <- as.factor(mtcars$vs)
gp <-ggplot(aes(x=wt, y=mpg, color=vs), 
             data=mtcars)
gp + geom_point()
```

#### Changing the Markers (the dots)

Here we are adding thea <code>aes()</code> function in the <code>geom_point()</code> function. In the <code>aes()</code> function we are adding the <em>color</em> and <em>shape</em> arguments and add the class column (the categorical variable).

```r
data(Burt, package = 'carData')
Burt$class <- as.factor(Burt$class)

gp <- ggplot(aes(x = IQbio, y = IQfoster), data = Burt)
gp + geom_point(aes(color = class, 
             shape = class))
```

### <span class="ez-toc-section" id="How_to_Add_a_Trend_Line_to_a_Scatter_Plot_in_R">How to Add a Trend Line to a Scatter Plot in R</span>

We use the <code>geom_smooth()</code> function and the method “lm” to add a regression line.

```r
gp <- ggplot(aes(x = IQbio, y = IQfoster), data = Burt)
gp + geom_point(aes(color = class, 
             shape = class)) +
      geom_smooth(method = "lm", se = FALSE)
```

In the next scatter plot example, we are going to add a regression line to the plot for each factor (category) also. Remember, we just add the color and shape arguments to the <code>geom_point()</code> function:

```r
gp + geom_point(aes(color = class, 
             shape = class)) +
      geom_smooth(aes(color = class), method = "lm", se = FALSE)
```

#### Bivariate Distribution on a Scatter plot

We are adding a bivariate distribution on the scatter plot in R using the <code>geom_density2d()</code> function.

```r
gp <- ggplot(aes(x=wt, y=mpg), 
             data=mtcars)
gp + geom_point() + geom_density2d()
```

### <span class="ez-toc-section" id="How_to_Add_Text_to_Scatter_Plot_in_R">How to Add Text to Scatter Plot in R</span>

Let's carry out correlation analysis using R, extract the <em>r</em>– and <em>p</em>-values, and later learn how to add this as text to our scatter plot. 

```r
require(dplyr)
require(broom)

corr <- mtcars  %$%
             cor.test(mpg, wt) %>%
             tidy %>%
             mutate_if(is.numeric, round, 4)

corr
```

```r
text = paste0('r = ', corr$estimate, ', ',
             ifelse(corr$p.value <= 0,
                           'p < 0.05',
                   paste('p = ', corr$p.value))
             )

text
```

#### Adding Text to a Plot in R

We add text using the<code>annotate</code> function. 

```r
gp <- ggplot(aes(x = wt, y = mpg), 
             data = mtcars)
gp + geom_point() + geom_smooth(method = "lm", se = FALSE) +
    annotate('text',  x = 4.5, y = 35, label=text)
```

#### More Complex Correlation Test and Text Example

```r
require(tidyr)
require(purrr)

data(Burt, package = 'carData')


corr <- Burt %>% group_by(class) %>%
  nest() %>% 
  mutate(Cor = map(data, ~ cor.test(.$IQbio, .$IQfoster)),
         p   = map_dbl(Cor, 'p.value'),
         est = map_dbl(Cor, 'estimate')
             ) %>%
  mutate_if(is.numeric, round, 4) %>%
  select(class, p, est, Cor)

text <- corr %>%
  mutate(
        text = paste0('r = ', est, ', ',
             ifelse(p <= 0.01,
                           'p < 0.05',
                   paste('p = ', p))))


Burt$class <- as.factor(Burt$class)

gp <- ggplot(aes(x = IQbio, y = IQfoster), 
             data = Burt) 


corrp <- gp + geom_point(aes(color = class, 
             shape=class)) +
      geom_smooth(aes(color = class), method = "lm", se = FALSE) +
      geom_text(aes(x = 120, y = 137, color="high", 
                    label=subset(text, class == "high")$text)) +
      geom_text(aes(x = 118, y = 109, color="medium", 
                    label=subset(text, class == "medium")$text)) +
      geom_text(aes(x = 124, y = 103, color="low", 
                    label=subset(text, class == "low")$text))

corrp
```

### <span class="ez-toc-section" id="How_to_Rotate_the_Axis_using_Ggplot2">How to Rotate the Axis using Ggplot2</span>

Here's how to rotate the axis labels

```r
data(Salaries, package = "carData")
Salaries$rank <- as.factor(Salaries$rank)

gp <- ggplot(aes(x = salary, y = yrs.since.phd), 
             data = Salaries) + 
    geom_point(aes(color = rank, 
             shape = rank)) +
    geom_smooth(method = "lm") +
   scale_y_continuous(limits = c(0, 60)) +
   scale_x_continuous(limits = c(50000, 240000), 
                      breaks = seq(50000, 240000, by = 10000))
```

To rotate the axis do this:

```r
gp + theme(axis.text.x = 
           element_text(angle = 90, hjust = 1))
```

### <span class="ez-toc-section" id="How_to_Style_a_Scatter_plot_in_R">How to Style a Scatter plot in R</span>

Here we use the <code>theme_bw()</code> function to get a dark-light themed plot. Then, we are going to make the scatter plot in black and grey colors using the <code>scale_colour_grey()</code> function. Finally, we add a theme layer using the function <code>theme()</code>. 

The function <code>element_blank()</code> will make draw “nothing” at that particular parameter. For instance, <code>plot.background = element_blank()</code> will give the plot a blank (white) background. 

```r
corrp + theme_bw() +  scale_colour_grey() +
  theme(axis.line = element_line(colour = "black")
        ,plot.background = element_blank()
        ,panel.grid.major = element_blank()
        ,panel.grid.minor = element_blank()
        ,strip.background = element_blank()
        ,panel.border = element_blank() 
        ,legend.title=element_blank()
        ,legend.key = element_blank())
```

### <span class="ez-toc-section" id="Saving_a_High_Resolution_Plot_in_R">Saving a High Resolution Plot in R</span>

In this section, we are going to learn how to save ggplot2 plots as PDF and TIFF files.

```r
data(Salaries, package = "carData")


gp <- ggplot(aes(x=yrs.since.phd, y=salary), 
             data=Salaries) + 
    geom_point() +
    geom_smooth(method = "lm", se = FALSE, colour="gray") +
    theme_bw() +
    theme(axis.line = element_line(colour = "black")
        ,plot.background = element_blank()
        ,panel.grid.major = element_blank()
        ,panel.grid.minor = element_blank()
        ,strip.background = element_blank()
        ,panel.border = element_blank() 
        ,legend.title=element_blank()
        ,legend.key = element_blank())  +
    xlab('Years since Ph.D.') +
    ylab('Salary')
```

Now we can use the <code>ggsave()</code> function to save the scatter plot.
